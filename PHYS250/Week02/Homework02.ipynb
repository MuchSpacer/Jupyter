```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
import scipy.interpolate as interp
```

```python
distances = np.array([-3.4083885, -5.6100975, -5.8256509, -3.9787359,  5.3127362, 5.9562532
, 4.4910645])
dates = np.array([1, 2, 3, 4, 7, 8, 9])
```

```python
dates[:4]
```

```python
dates[4:]
```

```python
lgr_min = interp.lagrange(dates[:4], distances[:4])
lgr_max = interp.lagrange(dates[4:], distances[4:])
```

```python
dates[4:]
```

```python
testdates = np.linspace(0, 10, 200)

plt.scatter(dates, distances)
plt.plot(testdates, lgr_min(testdates), label = "min")
plt.plot(testdates, lgr_max(testdates), label = "max")
plt.legend()
plt.ylim(-7, 7)
```

```python
drv_min = lgr_min.deriv()
drv_max = lgr_max.deriv()
```

```python
plt.plot(testdates, drv_min(testdates), label = "min")
plt.plot(testdates, drv_max(testdates), label = "max")
plt.legend()
plt.ylim(-7, 7)
```

```python
tmin = opt.brentq(drv_min, 2, 3)
tmin
```

```python
tmax = opt.brentq(drv_max, 7, 8)
tmax
```

```python
lgr_min(tmin)
```

```python
lgr_max(tmax)
```
