```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
```

```python
gcDat = np.loadtxt("gc.dat", skiprows = 2, unpack = True)
ls = gcDat[0][gcDat[4] != 999]
Vrs = gcDat[4][gcDat[4] != 999]
```

Preparing to fit our data using the model:

$V_r = V_{c,proj}\sin{l}$

via
$\begin{cases} V_r = V_{c,proj}l' \\ l' = \sin{l} \end{cases}$

```python
lprimes = np.sin(np.radians(ls))

def Vr_relationship(lprime, Vcproj):
    return Vcproj * lprime

Vcproj = opt.curve_fit(Vr_relationship, lprimes, Vrs)[0][0]
print(Vcproj)
```

```python
def Vr_new(l, Vcproj):
    return Vcproj * np.sin(np.radians(l))

Vcproj2 = opt.curve_fit(Vr_new, ls, Vrs)[0][0]
print(Vcproj2)
```

```python
smoothrange = np.linspace(0,360, 900, endpoint=False)
fitvalues = Vcproj * np.sin(np.radians(smoothrange))
```

```python
plt.scatter(ls, Vrs)
plt.plot(smoothrange, fitvalues, 'k--', label = "Fit: $-148\sin{\ell}$")
plt.grid()
plt.minorticks_on()
plt.tick_params(which = 'both', direction = 'in', top = True, right = True)
plt.title("Radial Velocity of Globular Clusters\nin the Milky Way Neighborhood")
plt.xlabel("Galactic Longitude $\ell$")
plt.ylabel("Radial Velocity $V_r$")
plt.legend()
plt.savefig("VrGraph.png")
```
