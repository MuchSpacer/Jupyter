```python
import numpy as np
import scipy.optimize as opt
import scipy.interpolate as itp
import scipy.signal as sig
import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rc('xtick', direction='out', top=True)
mpl.rc('ytick', direction='out', right=True)
mpl.rc('xtick.minor', visible=True)
mpl.rc('ytick.minor', visible=True)
plt.rcParams['axes.axisbelow'] = True
plt.rcParams['axes.xmargin'] = 0
```

```python
P, I = np.loadtxt("JE MM Interferometry 2024-10-17 15-57-04 AIR.csv", delimiter = ',', skiprows = 3)
```

```python
ind_pk, _ = sig.find_peaks(I)
ind_antipk, _ = sig.find_peaks(-I)
```

```python
plt.plot(P, I)
plt.scatter(P[ind_pk], I[ind_pk])
plt.scatter(P[ind_antipk], I[ind_antipk])
```

```python
plt.plot(P, I, marker = ".")
plt.xlabel("P, Torr")
plt.ylabel("LIA Output, V")
plt.title("Passing of Fringes: Air")
plt.savefig("air.png")
```

```python
Q = np.arange(1, 35.5, dtype = int)
```

```python
(m_pk, b_pk), cov = np.polyfit(P[ind_pk], Q, 1, cov = True)
sig_pk = np.sqrt(cov[0,0])
(m_antipk, b_antipk), cov = np.polyfit(P[ind_antipk], Q, 1, cov = True)
sig_antipk = np.sqrt(cov[0,0])
print(m_pk, m_antipk)
print(sig_pk, sig_antipk)
```

```python
x = np.linspace(110, 760, 50)
plt.plot(x, m_pk * x + b_pk)
plt.plot(x, m_antipk * x + b_antipk)
plt.scatter(P[ind_pk], Q, label = "Maxima")
plt.scatter(P[ind_antipk], Q, label = "Minima")
plt.ylim(bottom = 0)
plt.legend()
plt.xlabel("P, Torr")
plt.ylabel(r"Cumulative Number of Peaks $\Delta m$")
plt.title("Counting Fringes: Air")
plt.text(300, 5, rf"Maxima: $\Delta m = ({m_pk:.4f}\pm{sig_pk:.4f})P {b_pk:.4f}$"
         + "\n" + rf"Minima: $\Delta m = ({m_antipk:.4f}\pm{sig_antipk:.4f})P {b_antipk:.4f}$",
        bbox = dict(facecolor = "white"))
plt.savefig("airfit.png")
```

```python
P_He, I_He = np.loadtxt("JE MM Interferometry 2024-10-24 16-21-44 HE.csv", delimiter = ',', skiprows = 3)
```

```python
F = lambda P, A, f, phi, O : A * np.sin(2 * np.pi * f * P + phi) + O
Par, cov = opt.curve_fit(F, P_He, I_He, [0.13, 0.007, 0, 0.08])
f = Par[1]
sig_f = np.sqrt(cov[1, 1])
print(f, sig_f)
```

```python
plt.plot(P_He, I_He, label = "Helium Data", marker = ".")
X_He = np.linspace(min(P_He), max(P_He), 250)
plt.plot(X_He, F(X_He, *Par), "-.", label = "Sinusoidal Fit")
plt.legend()
plt.xlabel("P, Torr")
plt.ylabel("LIA Output, V")
plt.title("Passing of Fringes and Sinusoidal Fit: Helium") 
plt.savefig("helium.png")
```

```python
V_pz, I_pz, sig_pz = np.loadtxt("JE MM Interferometry 2024-10-15 16-16-32.csv", delimiter = ',', skiprows = 4)
```

```python
S = lambda V, offset, amp, A, B, phi: offset + amp * (np.sin(A * V **2 + B * V + phi) ** 2)
p0 = [0.18, -0.18, 0.01, 0.9, 1.3]
vals, cov = opt.curve_fit(S, V_pz, I_pz, p0)
```

```python
x_pz = np.linspace(0, 25, 200)
plt.plot(V_pz, I_pz, label = "Data", lw = 0.5)
plt.plot(x_pz, S(x_pz, *vals), "-.", label = "Quadratic Rate Fit")
plt.legend()
plt.xlabel("Input Voltage, V")
plt.ylabel("LIA Output, V")
plt.title("Passing of Fringes and Fit: Piezoelectric Slide")
plt.savefig("pzfit.png")
```
