```python
import scipy.special as sf
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.xmargin'] = 0
```

```python
P = lambda a, N: 1 - sf.erf(a*np.sqrt(N / 2))
```

```python
sampling_a = 1000
a0 = np.linspace(0,1,sampling_a)
a = np.full((3,sampling_a), a0)
N = np.array([1e2, 1e4, 1e6])
P_out = P(a.T, N)
```

```python
plt.semilogy(a0, P(a.T,N))
plt.grid()
plt.title(r"Probability of State: $|s| > \frac{N\alpha}{2}$ for Various $N$")
plt.ylabel("Probability")
plt.xlabel(r"Value of $\alpha$")
plt.text(0.01,1e-16,r"$10^6$")
plt.text(0.1,1e-16,r"$10^4$")
plt.text(0.85,1e-16,r"$N = 10^2$")
plt.savefig("HW1_01.png")
```
