## Thermal Conductivity Study (Final)

Using nine runs where the sample and stage were allowed to equilibriate while a steady current flowed through the heater. This gives a value for thermal conductance between the sample and stage, assuming a linear relationship between temperature difference and heat transfer rate.

However, the relationship is not a one-variable linear function. Possibly due to a calibration difference between the two thermistors, it is better represented with a two-variable function that accounts for both temperature difference and absolute temperature.

```python
import glob
import numpy as np
import scipy.optimize as opt
import scipy.signal as sig
import scipy.special as sf
import matplotlib.pyplot as plt
```

```python
V_set = np.array([3, 5, 9])
I_obs = np.array([11.518, 19.196, 34.538]) * 1e-3
P = V_set * I_obs
R = np.mean(V_set / I_obs)
```

```python
runs = glob.glob("./Conductivity/*.csv")

exclude = []

n = len(runs) - len(exclude)
```

```python
V = np.zeros(n)
Setpoints = np.zeros(n)
```

```python
times = []
stage_temps = []
sample_temps = []
i = 0
for r in runs:
    with open(r) as f:
        header = f.readline()
        
    elements = header.split()

    if(not((elements[0] + ' ' + elements[1]) in exclude)):
        V[i] = float(elements[4])
        Setpoints[i] = int(elements[2])
        
        time, stage, sample = np.loadtxt(r, delimiter = ',', skiprows = 1, unpack = True)
        times.append(time)
        stage_temps.append(stage)
        sample_temps.append(sample)
        i += 1
```

```python
t0 = 120 # time before heater is switched on
t1 = t0 + 120 # time before beginning fit (allows transients to die down)
F_eql = lambda t, A, tau, T_ss: A * np.exp(-(t-t1) / tau) + T_ss
```

```python
sample_ss = np.zeros(n)
diff_ss = np.zeros(n)
diff_ss_sig = np.zeros(n)
```

```python
for i in range(n):
    p0 = [-1, 200, 0]
    P, cov = opt.curve_fit(F_eql, times[i], sample_temps[i] - stage_temps[i], p0 = p0)
    _, _, diff_ss[i] = P
    diff_ss_sig[i] = np.sqrt(cov[2,2])
    p0 = [1, 200, 0]
    (_, _, sample_ss[i]), _ = opt.curve_fit(F_eql, times[i], sample_temps[i], p0 = p0)
```

```python
stage_ss = sample_ss - diff_ss
power = V ** 2 / R
```

```python
# Fitting function.
# x[0] is the temperature difference between stage and sample.
# x[1] is the stage temperature. alpha * x[1] is the stage temperature correction.
F_cond = lambda x, alpha, m, x0: m * (x[0,:] - alpha * x[1,:] + x0)
```

```python
xdata = np.array([diff_ss, stage_ss])

P, cov = opt.curve_fit(F_cond, xdata, power)
sig = np.sqrt(np.diag(cov))
```

```python
plt.scatter(diff_ss, power, c = stage_ss, cmap = "viridis")
X = np.array([np.linspace(0, 8), np.full(50, np.mean(stage_ss))])
plt.plot(X[0,:], F_cond(X, *P))
plt.ylabel("Heating Power")
plt.xlabel("Equilibrium Temperature Difference")
```

```python
print(f"Heat Transfer Rate: {P[1]:.3f} pm {sig[1]:.3f} W/degC")
print(f"Correction 1: {P[0]:.3f} pm {sig[0]:.3f} * Stage Temp degC")
print(f"Correction 2: {P[2]:.3f} pm {sig[2]:.3f} degC")
```
