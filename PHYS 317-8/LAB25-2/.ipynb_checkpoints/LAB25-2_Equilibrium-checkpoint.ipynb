## Thermal Conductivity Study (Preliminary)

Using a single run where the sample and stage were allowed to equilibriate while a steady current flowed through the heater. This gives a value for thermal conductivity between the sample and stage, assuming a linear relationship between temperature difference and heat transfer rate, and assuming the thermistors both have the same calibration. This should be taken to be an estimate.

```python
import numpy as np
import scipy.optimize as opt
import scipy.signal as sig
import scipy.special as sf
import matplotlib.pyplot as plt
```

```python
# Current through the heater was not digitally recorded.
# However, I observed it throughout the run to get an error estimate.
I_obs = 34.5e-3 # Amps pm .1 mA

# We are not automatically recording the voltage. The Agilent ARB function generator is set to 9v.
# Error in this can be pulled from the instrument's manual. (It should be negligible)
V_obs = 9
```

```python
t, T_sample, T_stage = np.loadtxt("Day2Eql.csv", unpack = True, delimiter = ',')
```

```python
plt.plot(t, T_sample)
plt.plot(t, T_stage)
```

```python
# Cooling cycle time was 120 s. Finding first point after the end of the cooling cycle.
i1 = 0
while t[i1] < 120:
    i1 = i1 + 1
print(i1)
```

```python
# adding on another 120 s to avoid transient effects
t2 = 240
i2 = 0
while t[i2] < t2:
    i2 = i2 + 1
print(i2)
```

```python
plt.plot(t[i2:], T_sample[i2:])
plt.plot(t[i2:], T_stage[i2:])
```

```python
F_eql = lambda t, A, a, C: A * np.exp(-(t-t2) / a) + C
```

```python
p0 = [.1, 100, 10]
P_stage_eql, cov1 = opt.curve_fit(F_eql, t[i2:], T_stage[i2:], p0)
P_sample_eql, cov2 = opt.curve_fit(F_eql, t[i2:], T_sample[i2:], p0)

print(P_stage_eql, np.sqrt(np.diag(cov1)))
print(P_sample_eql, np.sqrt(np.diag(cov2)))
```

```python
P_diff, cov = opt.curve_fit(F_eql, t[i2:], T_sample[i2:] - T_stage[i2:], p0)
print(P_diff, np.sqrt(np.diag(cov)))
```

```python
# Quick plotting fits to the sample temperature (above) and stage temperature (below)
X = np.linspace(t2, 1080, 200)
fig, [ax1, ax2] = plt.subplots(2, 1)
ax1.plot(X, F_eql(X, *P_sample_eql), "--")
ax1.plot(t[i2:], T_sample[i2:])
ax2.plot(X, F_eql(X, *P_stage_eql), "--")
ax2.plot(t[i2:], T_stage[i2:])
```

```python
# Quick plotting the fit to the difference between stage and sample temperature.
plt.plot(X, F_eql(X, *P_diff), "--")
plt.plot(t[i2:], T_sample[i2:] - T_stage[i2:])
```

```python
# The methodology here might not be 100% sound, since the stage temperature was still changing (no steady state).
# However, I still believe the asymptotic value of the difference is the best approximation I have for the steady state condition.

sigquad = np.sqrt(cov1[2,2] + cov2[2,2]) # errors added in quadrature
diff = P_sample_eql[2] - P_stage_eql[2]
print(f"Difference of Asymptotic Values: {diff:.4f} pm {sigquad:.4f}")
print(f"Asymptotic Value of Difference: {P_diff[2]:.4f} pm {np.sqrt(cov[2,2]):.4f}")
```

```python
P = V_obs * I_obs # Watts
K = P / P_diff[2] # Watts per K
print(f"Conductance: {K:.3f} W/K")
```
