```python
import numpy as np
import scipy.interpolate as interp
import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rc('xtick', direction='in', top=True)
mpl.rc('ytick', direction='in', right=True)
mpl.rc('xtick.minor', visible=True)
mpl.rc('ytick.minor', visible=True)
```

```python
period=365.25636;
e=0.01671123;
omega=2*np.pi/period;
```

```python
def t_theta_func(theta, e, period):
    '''Function gives the time taken by a body in orbit to sweep out an angle theta.
    t = 0, theta = 0 are chosen to be at periapsis.'''
    	
    time = period/(2*np.pi) * (2*np.arctan2(((1-e)/(1+e))**0.5*np.sin(theta/2),np.cos(theta/2))-(e*(1-e**2)**0.5*np.sin(theta))/(1+e*np.cos(theta)))
    return time 
```

```python
theta=np.deg2rad(np.arange(0,360.1,1));
spline = interp.InterpolatedUnivariateSpline(t_theta_func(theta,e,period), theta);
tgrid=np.linspace(0,period,5000)
ang_speed=spline(tgrid,nu=1)
```

```python
def kepler_t(psi, e, period):
    time = (psi-e*np.sin(psi))*period/(2*np.pi)    
    return time

psi=np.deg2rad(np.arange(0,180.1,1))
time=kepler_t(psi,e,period)
```

```python
def theta_psi(psi, e):
    theta = np.arccos((np.cos(psi)-e)/(1-e*np.cos(psi)))
    return theta
```

```python
spline2 = interp.InterpolatedUnivariateSpline(kepler_t(psi, e, period),theta_psi(psi, e))
ang_speed_2 = np.zeros_like(tgrid)
N = len(ang_speed_2)
ang_speed_2[:N//2] = spline2(tgrid[:N//2],1)
ang_speed_2[N//2:] = spline2(tgrid[:N//2],1)[::-1]
```

```python
plt.plot(tgrid, spline(tgrid))
```

```python
plt.plot(tgrid, ang_speed)
```

```python
plt.plot(tgrid, ang_speed_2)
```

```python
abs_diff = np.abs(np.degrees(ang_speed - ang_speed_2))
plt.semilogy(tgrid, abs_diff, "r,")

plt.title("Comparison of Methods: Orbit Equation vs. Kepler's Equation")
plt.xlabel("Time Since Periapsis $t$, days")
plt.ylabel("Absolute Difference $\\Delta\\theta$, degrees")
plt.grid(which = "both", axis = "both", alpha = 0.25)
```

```python
bins = np.logspace(-14, -8, 100)

plt.hist(abs_diff, bins)

plt.semilogx()

plt.xlabel("Absolute Error")

plt.ylabel("Number of Occurrences")
```
