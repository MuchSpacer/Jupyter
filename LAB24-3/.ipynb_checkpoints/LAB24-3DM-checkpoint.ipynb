```python
def driven(z, l, r, prom = 0.00005):
    x = 0.9428665054936394 # see FD notebook.
    zs = z[l:r]
    ind_pk, _ = sig.find_peaks(zs, prominence = prom)
    ind_antipk, _ = sig.find_peaks(-zs, prominence = prom)
    l = min(len(ind_pk), len(ind_antipk))
    a = (1 - x) * (np.sum(zs[ind_pk[:l]]) - np.sum(zs[ind_antipk[:l - 1]]))
    b = - zs[ind_pk[0]] + x * zs[ind_pk[l-1]]
    c = (2 * l + 1) * (1 + x)
    return (a + b) / c
```

```python
def Gcalc(K, D, d = 66.5e-3, R = 44.76e-3, M = 874.35e-3, m = 14.9515e-3, mh = .34e-3, mb = 6.5576e-3, fb = 0.19, fd = 0.035):
        num = K * D * R**2
        denom = 2 * M * ((m - mh) * (1-fd) + mb * fd) * d
        return num / denom
```
