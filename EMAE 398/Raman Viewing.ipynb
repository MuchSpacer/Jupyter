```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as itp
from RamanSpectrum import *
```

```python
plt.rcParams['xtick.top'] = True
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['xtick.minor.visible'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.minor.visible'] = True
plt.rcParams['axes.axisbelow'] = True
plt.rcParams['axes.xmargin'] = 0
plt.rcParams['axes.grid'] = True
plt.rcParams['axes.grid.which'] = "both"
```

```python
## Parameters ##
L_cutoff = 300
R_cutoff = 1900

same_axes = False
```

```python
files_to_display = []

titles = []

with open("files_to_display.txt") as f:
    files_to_display = f.readlines()
    for i in range(len(files_to_display)):
        files_to_display[i] = files_to_display[i].rstrip()
        titles.append((files_to_display[i].split('/'))[-1])

spectra = []

for f in files_to_display:
    spectra.append(RamanSpectrum(f, L_cutoff, R_cutoff))
```

```python
if same_axes:
    n = 1
else:
    n = len(spectra)

fig, axs = plt.subplots(n, 2, sharex = True)

if len(np.shape(axs)) == 1:
    axs = (axs,)

i = 0
for spectrum in spectra:
    spectrum.plot_raw(axs[i][0])
    axs[i][0].set_title(titles[i])
    spectrum.plot_peaks(axs[i][1])
    axs[i][1].set_title(titles[i])
    if not same_axes:
        i += 1
```
