```python
import glob
import numpy as np
import scipy.optimize as opt
import scipy.signal as sig
import scipy.special as sf
import matplotlib.pyplot as plt
```

## Data

During each run, resistance of the sample and stage thermistors was gathered using the digital-programmable Keithley 2000 multimeters. This was converted to temperature using a spline. Instruments were controlled and data recorded using a MATLAB script. The first line of each `.csv` file is a header including the date and time of the experiment, the setpoint thermistor resistance, and the equilibriation time and heating time in seconds.

The first column of data is time and the second is temperature in Celsius.

```python
runs = glob.glob("./Runs/*.csv")


# Runs where the sample temperature was still changing when the heating cycle began
# are listed so they can be included or excluded during analysis.
non_equilibrium = ["2025-03-06 16-16-54",
                   "2025-03-06 16-41-58",
                   "2025-03-18 16-08-03",
                   "2025-03-18 17-18-14",
                   "2025-03-18 14-29-42",
                  ]

# Runs subject to major experimental problems can be excluded by date and time.
exclude = ["2025-03-18 13-28-17", # Drift
           "2025-03-18 13-46-52", # Drift
           "2025-03-18 16-45-25", # Refrigerator driven
           "2025-03-18 17-06-00", # Hysteresis?
           "2025-03-20 15-17-26", # step change
           "2025-03-20 16-03-30", # Refrigerator driven (might be okay)
           "2025-03-27 14-37-32", # Stage went wild
          ] #+ non_equilibrium

n = len(runs) - len(exclude)

# list to be populated with paths of all included runs
included = []
```

```python
t1 = np.zeros(n) # Array of equlilibriation times of each run
t2 = np.zeros(n) # Array of heating times of each run
Setpoints = np.zeros(n) # Array of thermistor resistance setpoints for each run
```

```python
times = []
stage_temps = []
sample_temps = []
i = 0
for r in runs:
    with open(r) as f:
        header = f.readline()
        
    elements = header.split()

    if(not((elements[0] + ' ' + elements[1]) in exclude)):
        included.append(r)        
        
        t1[i] = int(elements[4])
        t2[i] = int(elements[6])
        Setpoints[i] = int(elements[2])
        
        time, stage, sample = np.loadtxt(r, delimiter = ',', skiprows = 1, unpack = True)
        times.append(time)
        stage_temps.append(stage)
        sample_temps.append(sample)
        i += 1
```

## Parameters

```python
# Voltage setpoint to drive the heater + observed current.
V = 9 # Volts
I = 34.54e-3 # Amperes

# Thermal conductance determined per results in LAB25-2_Equilibrium.ipynb
thermal_conductance = 0.0464 # W/deg. C
```

## Analysis

```python
# the difference between sample and stage temperature.
# per results in LAB25-2_Equilibrium.ipynb:
# temperature differences can be corrected by -0.833 C -0.083 * stage temp (in C).
# This accounts for heat flow from the environment and calibration differences.
# Although, after some examination, this 'correction' doesn't seem to be very accurate.
sample_stage_deltas = []
sample_stage_deltas_corr = []


for i in range(n):
    delta_corr = (sample_temps[i] - (stage_temps[i] * 1.083) - 0.833)
    sample_stage_deltas.append(sample_temps[i] - stage_temps[i])
    sample_stage_deltas_corr.append(delta_corr)
```

```python
# Fitting for relaxation times.

F = lambda t, t0, A, Tf, tau: Tf + A * np.exp(-(t - t0) / tau)

delay = 90 # additional time to wait after heating ends for transient effects to die down.

A = np.zeros(n) # only recorded for the production of graphs.
tau = np.zeros(n)
tau_sig = np.zeros(n)
T_ss = np.zeros(n)
T_ss_sig = np.zeros(n)

for i in range(n):
    t0 = t1[i] + t2[i] + delay
    F_prime = lambda t, A, Tf, tau: F(t, t0, A, Tf, tau)
    indices = times[i] > t0
    p0 = [1, sample_temps[i][-1], 150] # initial guess for parameters
    P, cov = opt.curve_fit(F_prime, times[i][indices], sample_temps[i][indices], p0 = p0)
    A[i], T_ss[i], tau[i] = P
    _, T_ss_sig[i], tau_sig[i] = np.sqrt(np.diag(cov))
```

```python
# Fitting for relaxation times with stage temperature subtracted.
# For slow changes in the stage temperature, the rate of heat transfer
# can be pretty accurately related to the instantaneous temperature difference.
# So hopefully this will provide more accurate analysis of non-equilibrium runs.

A_ = np.zeros(n) # only recorded for the production of graphs.
tau_ = np.zeros(n)
tau_sig_ = np.zeros(n)
T_ss_ = np.zeros(n)
T_ss_sig_ = np.zeros(n)

for i in range(n):
    t0 = t1[i] + t2[i] + delay
    F_prime = lambda t, A_, Tf, tau_: F(t, t0, A_, Tf, tau_)
    indices = times[i] > t0
    p0 = [1, sample_temps[i][-1], 150] # initial guess for parameters
    P, cov = opt.curve_fit(F_prime, times[i][indices], sample_stage_deltas[i][indices], p0 = p0)
    A_[i], T_ss_[i], tau_[i] = P
    _, T_ss_sig_[i], tau_sig_[i] = np.sqrt(np.diag(cov))
```

```python
# Finding the change in temperature caused by the heat input.

# percent of electrical power converted to heat in the sample
conversion_rate = 0.95

# Heat input for each sample (power times time)
Q_in = V * I * t2 * conversion_rate

delta_Ts = np.zeros(n)
Q_out = np.zeros(n)
Q_net = np.zeros(n)

j1s = np.zeros(n) # heating start indices
j2s = np.zeros(n) #
j3s = np.zeros(n)

delta_T12s = np.zeros(n)

# Finding instantaneous change in temperature after heating for each sample.
# Plus finding/recording the total heat flow out.
for i in range(n):
    # Index at which heating begins.
    j1 = np.argmax(times[i] > t1[i])
    j1s[i] = j1
    T1 = sample_temps[i][j1]

    j2 = np.argmax(times[i] > t1[i] + t2[i])
    j2s[i] = j2
    T2 = sample_temps[i][j2]

    delta_T12s[i] = T2 - T1
    
    j3 = np.argmax(sample_temps[i][j1:]) + j1
    j3s[i] = j3
    T3 = sample_temps[i][j3]

    delta_Ts[i] = T3 - T1
    
    T_diff_from_ss = sample_stage_deltas[i][j1:j2] - T_ss_[i]

    dt = times[i][j1:j2] - times[i][j1-1:j2-1]
    dQ = dt * T_diff_from_ss * thermal_conductance
    Q_out[i] = np.sum(dQ)
    
Q_net = Q_in - Q_out
```

```python
# Heat capacities found via the instantaneous temperature change method.
C_inst = Q_net / delta_Ts

# Heat capacities found via relaxation time.
C_relax = tau_ * thermal_conductance
```

```python
quadratic_fit = lambda T, a, R0, R1: a * (T - R0) * (T - R1)
```

```python
p0 = [-1, 0, 30]
p_q_relax, cov_q_relax = opt.curve_fit(quadratic_fit, T_ss, C_relax, p0 = p0)
p_q_inst, cov_q_inst = opt.curve_fit(quadratic_fit, T_ss, C_inst, p0 = p0)
```

```python
T_pk_relax = abs(p_q_relax[2] + p_q_relax[1]) / 2
std_pk_relax = np.sqrt(cov_q_relax[1,1] + cov_q_relax[2,2])

T_pk_inst = abs(p_q_inst[2] + p_q_inst[1]) / 2
std_pk_inst = np.sqrt(cov_q_inst[1,1] + cov_q_inst[2,2])

print(f"Peak - Relaxation Time Method: {T_pk_relax:.1f} pm {std_pk_relax:.1f} C")
print(f"Value: {quadratic_fit(T_pk_relax, *p_q_relax):.1f} J / K")
print(f"Peak - Instantaneous Rise Method: {T_pk_inst:.1f} pm {std_pk_inst:.1f} C")
print(f"Value: {quadratic_fit(T_pk_inst, *p_q_inst):.1f} J / K")
```

```python
linear_fit = lambda T, a, T0: (a * T) + T0
```

```python
p0 = [0, 5]
p_l_relax, cov_l_relax = opt.curve_fit(linear_fit, T_ss, C_relax, p0 = p0)
p_l_inst, cov_l_inst = opt.curve_fit(linear_fit, T_ss, C_inst, p0 = p0)
print(p_l_relax)
print(p_l_inst)

print(f"Slope - Relaxation Time Method: {p_l_relax[0]:.2f} pm {np.sqrt(cov_l_relax[0,0]):.2f}")
print(f"T0 - Relaxation Time Method: {p_l_relax[1]:.2f} pm {np.sqrt(cov_l_relax[1,1]):.2f}")

print(f"Slope - Instantaneous Rise Method: {p_l_inst[0]:.2f} pm {np.sqrt(cov_l_inst[0,0]):.2f}")
print(f"T0 - Instantaneous Rise Method: {p_l_inst[1]:.2f} pm {np.sqrt(cov_l_inst[1,1]):.2f}")
```

```python
print(f"Typical temp difference at sensor after heating: {np.mean(delta_T12s):.3f} C")
```

```python
print(f"Average temp. change overall: {np.mean(delta_Ts):.3f}")
```

```python
vol = 4.8
vol25_relax = linear_fit(25, *p_l_relax) / vol
vol25_inst = linear_fit(25, *p_l_inst) / vol

print(f"Volume heat capacity - Relaxation Time Method: {vol25_relax:.2f}")
print(f"Volume heat capacity - Instantaneous Rise Method: {vol25_inst:.2f}")
```
