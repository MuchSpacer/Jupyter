0.6 * m + K = log N

set equal at m = 4.0 (or rather, m = NmDat[0][0]

--> K = log N - 0.6 * 4.0

```python
constK = NmDat[2][0] - 0.6 * NmDat[0][0]

plt.axes().set_facecolor("black")
plt.plot(NmDat[0], 0.6 * NmDat[0] + constK, ls = "--", color = "white", label = "Uniform Infinite Distribution") # Expected Relationship
plt.plot(NmDat[0], NmDat[2], marker = "s", c = "xkcd:medium blue", markersize = 6., label = "Stars in Galactic Plane") # Actual Relationship - in galactic plane
plt.plot(NmDat[0], NmDat[1], marker = "D", c = "orange", markersize = 6., label = "Stars out of Galactic Plane") # Actual Relationship - out of galactic plane
plt.xlabel("Apparent Magnitude $m$", size = "large")
plt.ylabel("Star Count $\\log N(m)$", size = "large")
plt.title("Star Counts Observed from Earth")
plt.legend()
plt.minorticks_on()
plt.grid(visible = True, which = "major")
plt.grid(visible = True, which = "minor", alpha = 0.5)
```

### Explanation

Both star counts increase more slowly than would be expected in an infinite, uniform universe, and they do not follow a linear trend. Additionally, the $\\log(N)$ count of stars outside the galactic plane is offset vertically from the in-plane count, meaning some fraction fewer stars of all magnitudes are visible.

Both of these trends are explained by the shape of the Milky Way galaxy. Because the galaxy is not infinite in size, eventually the edge of the galaxy is reached and the number of stars observed decreases. Additionally, the galaxy's flat shape means that far fewer stars exist outside of the disk than within it.

### Fun With Magnitudes

$M = -2.5\times\log{L} + K$

$\log{NL} = \log{N} + log{L}$

$M_N = M_0 - 2.5\log{N}$

```python
def MagCalculator(Ms, Ns, BVs):
    Ms = np.array(Ms)
    Ns = np.array(Ns)
    BVs = np.array(BVs)
    def magAdd(M, N):
        return M - 2.5 * np.log10(N)
    Bs = Ms + BVs
    Mts = magAdd(Ms, Ns)
    Bts = magAdd(Bs, Ns)
    Mt = -2.5 * np.log10(np.sum(10 ** (-0.4 * Mts)))
    Bt = -2.5 * np.log10(np.sum(10 ** (-0.4 * Bts)))
    return (Mt, Bt)

def mA(M, N):
        return M - 2.5 * np.log10(N)
```

```python
sunCluster = MagCalculator(4.83, 1e6, 0.65)
print(f"V-band Magnitude: {sunCluster[0]}")
print(f"New B-V Color: {sunCluster[1]- sunCluster[0]}")
```

$M_0 - M_1 = -2.5 \times (\log{L_0} - \log{L_1}) = -2.5 \times \log{L_0 / L_1}$

$-0.4(M_0 - M_1) = \log{L_0 / L_1}$



```python
mixedCluster = MagCalculator([1.0, 4.83], [1e5, 1e6], [1.0, 0.65])
justGiants = MagCalculator(1.0, 1e5, 1.0)
print(f"V-band Magnitude: {mixedCluster[0]}")
print(f"New B-V Color: {mixedCluster[1]- mixedCluster[0]}")
print(f"Fraction of red giant contribution: {10 ** (-0.4 * (justGiants[0] - mixedCluster[0]))}")
```

### Galactic Coordinates

$z = d\cos{l}\cos{b}$

```python
def galZ(b, d):
    return d * np.sin(np.radians(b))

print(f"1. M13: {galZ(40.9, 7000)} pc.")
print(f"2. Orion Nebula: {galZ(-19.4, 450)} pc.")
```

Neither M13 nor Orion are in the direction of the galactic center. M13 has too much vertical distance from earth to be in the galactic disk, while Orion is without a doubt in the disk. Therefore I would say that M13 is in the halo, while Orion is in the disk, but I cannot say whether it is in a spiral arm.

```python
def galcoords(l, b, d, degrees = True):
    if degrees:
        l = np.degrees(l)
        b = np.degrees(b)
    x = d * np.cos(l) * np.cos(b)
    y = d * np.sin(l) * np.cos(b)
    z = d * np.sin(b)
    return (x, y, z)
    # When run on a numpy array, will return a triple of arrays.
```

### Globular Clusters

```python
gcDat = np.loadtxt("gc.dat", skiprows = 2, unpack = True)
gcXYZ = galcoords(gcDat[0], gcDat[1], gcDat[2])
metalPoor = (gcDat[3] != 999.) & (gcDat[3] < -0.8)
metalRich = (gcDat[3] != 999.) & (gcDat[3] >= -0.8)
```

```python
fig, (xy,xz) = plt.subplots(2, 1, sharex = "all", sharey = "all")

plt.ylim(-60, 60)
xy.minorticks_on()
xz.minorticks_on()

for axis in fig.axes:
    axis.set_facecolor("black")
    axis.set_aspect("equal")
    axis.grid(True, "both", "both", alpha = 0.5)
    axis.set_xlim(-60, 60)
    axis.axhline(c = "white")
    axis.axvline(c = "white")
    
fig.suptitle("Locations of Globular Clusters relative to Earth")
xz.set_xlabel("$X$ (kpc) Towards Galactic Center")
xy.set_ylabel("$Y$ (kpc)")
xz.set_ylabel("$Z$ (kpc)")

commonSize = 9

xy.scatter(gcXYZ[0][metalRich], gcXYZ[1][metalRich], color = "xkcd:sky", s = commonSize, zorder = 100, label = "Metal-Rich")
xy.scatter(gcXYZ[0][metalPoor], gcXYZ[1][metalPoor], marker = "s", color = "xkcd:light red", s = commonSize, zorder = 90, label = "Metal-Poor")

xy.legend(loc = "upper left")

xz.scatter(gcXYZ[0][metalRich], gcXYZ[2][metalRich], color = "xkcd:sky", s = commonSize, zorder = 100,  label = "Metal-Rich")
xz.scatter(gcXYZ[0][metalPoor], gcXYZ[2][metalPoor], marker = "s", color = "xkcd:light red", s = commonSize, zorder = 90, label = "Metal-Poor")

xz.legend(loc = "upper left")

fig.set_size_inches((8, 10))
plt.tight_layout()
```

#### Comparing Metal-rich and Metal-poor Distributions

The metal-rich clusters are much more tightly distributed in both Y and Z, whereas some metal-poor clusters are 60 kpc or more from Earth, firmly outside of the galaxy. There also appear to be slightly more metal-rich clusters in the positive X direction (towards the galactic center). Both of these effects might be caused by the origins of these clusters. Metal-poor clusters are much older and may have formed outside of the galaxy, or been ejected from it. Metal-rich clusters must form in metal-rich regions, those where generations of stars have already been born. Therefore, the metal-rich clusters likely formed within the galaxy.

#### Finding the Galactic Center

I would expect metal-rich clusters to give a better idea of the shape and location of the galactic center because they are younger and must have formed within the galaxy.

```python
(xMR, yMR, zMR) = (np.mean(gcXYZ[0][metalRich]),np.mean(gcXYZ[1][metalRich]),np.mean(gcXYZ[2][metalRich]))
(xMP, yMP, zMP) = (np.mean(gcXYZ[0][metalPoor]),np.mean(gcXYZ[1][metalPoor]),np.mean(gcXYZ[2][metalPoor]))
```

```python
print(f"Metal-rich center in X, Y, Z: {(xMR, yMR, zMR)}")
print(f"Metal-poor center in X, Y, Z: {(xMP, yMP, zMP)}")
```

```python
print(f"Distance to metal-rich center: {np.sqrt(xMR**2 + yMR**2 + zMR**2)}")
print(f"Distance to metal-poor center: {np.sqrt(xMP**2 + yMP**2 + zMP**2)}")
```

#### Result

According to this data, using the center of the metal-rich distribution, the galactic center is about 1.8 kpc away. I know that the galactic center is on the order of 10kpc away. It is likely that observational bias factored into this data, because it would be much more difficult to observe clusters on the opposite side of the galactic center.
