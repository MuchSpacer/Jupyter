## Problem 1

```python
def Q1f(x):
    return np.exp(x * -1/3)

def Q1f_drv(x):
    return -1/3 * Q1f(x)

def cdiff(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)
    
def fdiff(f, x, h):
    return (f(x + h) - f(x)) / h

def efrac(val, expected):
    return abs((val - expected) / expected)
```

```python
Q1x = 0.7

Q1_steps = np.logspace(0, -15, 16)

Q1_expect = Q1f_drv(Q1x)

Q1_cdiffs = cdiff(Q1f, Q1x, Q1_steps)

Q1_errors = efrac(Q1_cdiffs, Q1_expect)
```

```python
plt.loglog(Q1_steps, Q1_errors)
plt.grid("both", "both", alpha=0.25)
plt.xlabel("Step Size")
plt.ylabel("Fractional Error")
plt.title("Center Differencing Fractional Error");
```

## Problem 2

```python
def Q2f(x):
    return np.exp2(x) * np.cos(x)

Q2_ln2 = np.log(2)

def Q2f_deriv(x):
    term1 = np.exp2(x) * -np.sin(x)
    term2 = Q2_ln2 * np.exp2(x) * np.cos(x)
    return term1 + term2
```

```python
def richardson_center (f, z, h, nsteps, args=()):
    """Evaluate the first derivative of a function at z, that is f'(z),
    using Richardson extrapolation and center differencing.

    Returned is the full table of approximations, Fij for j <= i. The
    values of Fij for j > i are set to zero. The final value F[-1,-1]
    should be the most accurate estimate.

    Parameters
    ----------
    f : function
        Vectorized Python function.
        This is the function for which we are estimating the derivative.
    z : number
        Value at which to evaluate the derivative.
    h : number
        Initial stepsize.
    nsteps : integer
        Number of steps to perform.
    args : tuple, optional
        extra arguments to pass to the function, f.
    """
    # Extra check to allow for args=(1) to be handled properly. This is a
    # technical detail that you do not need to worry about.
    if not isinstance(args, (tuple, list, np.ndarray)):
        args = (args,)
    # Create a zero filled table for our estimates
    F = np.zeros((nsteps, nsteps))
    # First column of F is the center differencing estimate.
    # We can fill this without a loop!
    harr = h / 2.**np.arange(nsteps)
    F[:,0] = (f(z+harr, *args) - f(z-harr, *args)) / (2.*harr)
    # Now iterate, unfortunately we do need one loop. We could
    # get rid of the inner loop but the algorithm is a little easier to
    # understand if we do not.
    for i in range(1, nsteps):
        fact = 0.25
        for j in range(1, i+1):
            F[i,j] = F[i-1,j-1] - (F[i-1,j-1] - F[i,j-1])/ (1-fact)
            fact *= 0.25
    return F
```

```python
def richardson_forward (f, z, h, nsteps, args=()):
    """Evaluate the first derivative of a function at z, that is f'(z),
    using Richardson extrapolation and forward differencing.

    Returned is the full table of approximations, Fij for j <= i. The
    values of Fij for j > i are set to zero. The final value F[-1,-1]
    should be the most accurate estimate.

    Parameters
    ----------
    f : function
        Vectorized Python function.
        This is the function for which we are estimating the derivative.
    z : number
        Value at which to evaluate the derivative.
    h : number
        Initial stepsize.
    nsteps : integer
        Number of steps to perform.
    args : tuple, optional
        extra arguments to pass to the function, f.
    """
    # Extra check to allow for args=(1) to be handled properly. This is a
    # technical detail that you do not need to worry about.
    if not isinstance(args, (tuple, list, np.ndarray)):
        args = (args,)
    # Create a zero filled table for our estimates
    F = np.zeros((nsteps, nsteps))
    # First column of F is the center differencing estimate.
    # We can fill this without a loop!
    harr = h / 2.**np.arange(nsteps)
    F[:,0] = (f(z+harr, *args) - f(z, *args))/ harr
    # Now iterate, unfortunately we do need one loop. We could
    # get rid of the inner loop but the algorithm is a little easier to
    # understand if we do not.
    for i in range(1, nsteps):
        fact = 0.5
        for j in range(1, i+1):
            F[i,j] = F[i-1,j-1] - (F[i-1,j-1] - F[i,j-1])/ (1-fact)
            fact *= 0.5
    return F
```

```python
Q2_expect = Q2f_deriv(1.15)
```

```python
Q2_rcM = richardson_center(Q2f, 1.15, 0.4, 8)
Q2_rc = Q2_rcM[-1][-1]
print(Q2_expect)
print(Q2_rc)
print(efrac(Q2_rc, Q2_expect))
```

```python
Q2_rfM = richardson_forward(Q2f, 1.15, 0.4, 8)
Q2_rf = Q2_rfM[-1][-1]
print(Q2_expect)
print(Q2_rf)
print(efrac(Q2_rf, Q2_expect))
```

```python
def iterFinder(rM, exp):
    count = 1
    results = np.diag(rM)
    for result in results:
        if efrac(result, exp) <= 1e-12:
            print(count)
            return
        count += 1
```

```python
Q2_rcLong =richardson_center(Q2f, 1.15, 0.4, 12)

iterFinder(Q2_rcLong, Q2_expect)
Q2_err_rc = efrac(np.diag(Q2_rcLong), Q2_expect)
```

```python
Q2_rfLong =richardson_forward(Q2f, 1.15, 0.4, 12)

iterFinder(Q2_rfLong, Q2_expect)
Q2_err_rf = efrac(np.diag(Q2_rfLong), Q2_expect)
```

```python
plt.semilogy(np.linspace(1,12,12), Q2_err_rc)
plt.semilogy(np.linspace(1,12,12), Q2_err_rf)
```
